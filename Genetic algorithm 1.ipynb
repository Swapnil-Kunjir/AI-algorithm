# minimization of ax^2+bx+c

```python
from random import randint
import collections
import itertools
```

```python
# for getting 1000 size population of each having 4 variables for a,b,c and x
z= [[randint(-50, 50) for p in range(4)] for p in range(1000)]
```

```python
# this is the equation that we want to minimize
def equation(z,i):
    q=z[i][0]*(z[i][3]**2)+z[i][1]*(z[i][3])+z[i][2]
    return q
```

### Finding top 10
for finding 10 fittest parents which we will use for further crossover and mutation.<br>
fittest i.e. population sample having lowest value for  equation q which will help us eventually to get close to 0.

```python

def ten_fittest_parents(z):
    z_dict={}
    for i in range(len(z)):
        z_dict.update({abs(equation(z,i)):z[i]})
    new=dict(sorted(z_dict.items()))    
    
    to_new=dict(itertools.islice(new.items(), 10))
    
    top_parents=list(to_new.values())
    return to_new,top_parents
    
```

### Crossover
crossover between 5 pairs of 2 parents each for which we will keep 2 values a and b same and exchange value of c and x of first parent with value of c and x of second parent. <br>
This will lead to us having 10 children or offspring along with 10 parents.

```python

def crossover(top_parents):
    couple=[((i),(i+1) % len(top_parents)) for i in range (len(top_parents))][0::2]
    
    offspring=[]
    for i in range(len(couple)):
        p1=top_parents[couple[i][0]]
        p2=top_parents[couple[i][1]]
        o1=[]
        o2=[]
        
        o1[0:2]=p1[0:2]
        o1[2:]=p2[2:]
        o2[0:2]=p2[0:2]
        o2[2:]=p1[2:]
        offspring.append(o1)
        offspring.append(o2)
    osp={}
    for i in range(len(offspring)):
        osp.update({abs(equation(offspring,i)):offspring[i]})
        
    cross_dict=dict(sorted(osp.items()))
    return offspring, cross_dict
    
```

### Mutation <br>
Here we will mutate children i.e. we will randomly change one of 4 values with some other

```python

def mutation(offspring):
    for i in range(len(offspring)):

        y=randint(0,3)
        
        offspring[i][y]=randint(-50,50)
    
    osp={}
    for i in range(len(offspring)):
        osp.update({abs(equation(offspring,i)):offspring[i]})
    mut_dict=dict(sorted(osp.items()))
    return offspring,mut_dict
```

```python
z= [[randint(-50, 50) for p in range(4)] for p in range(1000)]
i=0
while True:
        a='generation'+str(i)
        print(a)
        i=i+1
        print('original')
        
        to_new,top_parents=ten_fittest_parents(z)
        
        print(to_new)

        if (list(to_new.keys())[0]==0):
            print({list(to_new.keys())[0]:list(to_new.values())[0]})
            break
# If we found solution after zeroth generation the loop will break and show solution 
# or otherwise it will be result of mutation from 1st generation even if it shows original they are original parents for next gen.
        
        print('crossover')
        
        offspring,cross_dict=crossover(top_parents)
        
        print(cross_dict)
        
# If we found solution through crossover we break our loop and print solution

        if (list(cross_dict.keys())[0]==0):
            print({list(cross_dict.keys())[0]:list(cross_dict.values())[0]})
            break
        
        print('mutation')
        
# We take offsprings found in crossover for mutation if we haven't found solution

        offspring_mut,mut_dict=mutation(offspring)
        
        print(mut_dict)
        
#after mutation we take 10 parents and 10 children back to finding 10 fittest so they can be next generation parents
        
        z=top_parents+offspring_mut
        
```

## we can see we found answer in 18th generation through mutation
